```python
import pandas as pd

df = pd.read_csv('data/stocks_to_display.csv')
df
```

```python
df = df.drop('Unnamed: 0', 1)
df
```

```python
df = df.set_index('Date')
df
```

```python
df.plot(kind='line', y=['Adj Close IBM', 'Adj Close AAPL', 'Adj Close AMZN'])
```

```python
df[0:100].plot(kind='line', y=['Adj Close IBM', 'Adj Close AAPL', 'Adj Close AMZN'])
```

```python
df_display = df[['Adj Close IBM', 'Adj Close AAPL', 'Adj Close AMZN']]
df_display
```

```python
# lambda x: x / x[0]
def my_func(x):
    return x / x[0]

df_display = df_display.apply(my_func)
df_display
```

```python
df_display.plot(kind='line')
```

```python
df_display[0:1000].plot(kind='line')
```

```python
df_display.plot(kind='line', subplots=True)
```

```python
df_display.plot(kind='line', logy=True)
```

```python
df_display.plot(kind='line', title='Stock Prices')
```

```python
df_display.plot(kind='line', title='Stock Prices', logy=True, grid=True)
```

```python
df_aapl = pd.read_csv('data/AAPL.csv')
df_aapl
```

```python
df_aapl = df_aapl.set_index('Date')
df_aapl
```

```python
df_aapl['5 Day Moving Average'] = df_aapl['Adj Close'].rolling(5).mean()
df_aapl.head(10)
```

```python
df_aapl['200 Day Moving Average'] = df_aapl['Adj Close'].rolling(window=200).mean()
df_aapl
```

```python
df_aapl[['Adj Close', '200 Day Moving Average']].plot(kind='line', figsize=(17, 7), title='Apple (AAPL)')
df_aapl[['Volume']].plot(kind='area', color='red', figsize=(17, 3))
```
